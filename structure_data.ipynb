```python
import pandas as pd
```

```python
file_path = "trial3/unstructured_unencrypted_read3.csv" 
final_path = "trial3/unencrypted_read3.csv"
```

```python
df = pd.read_csv(file_path, sep='\t', header=0)
```

```python
df
```

```python
file_sizes = df.iloc[:,0]
```

```python
file_sizes
```

```python
block_sizes = df.columns[1:].astype(str)
```

```python
structured_data = []
```

```python
for i, file_size in enumerate(file_sizes):
    for j, block_size in enumerate(block_sizes):
        speed = df.iloc[i, j + 1]
        structured_data.append([file_size, block_size, speed])
```

```python
final_df = pd.DataFrame(structured_data, columns=["File Size", "Block Size", "Speed"])
```

```python
final_df
```

```python
final_df.to_csv(final_path)
```
