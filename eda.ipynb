```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = "data.csv"
df = pd.read_csv(file_path)
```

```python
sns.set_palette("colorblind")
```

```python
df
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Encryption", y="Speed", hue="Operation", data=df)
plt.title("Speed by Encryption and Operation")
plt.show()
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(2, 1, 1) 
sns.lineplot(x="Block Size", y="Speed", hue="Encryption", data=df[df["Operation"] == "read"], markers=True)
plt.title("Speed vs. Block Size (Read)")
plt.xlabel("Block Size")
plt.ylabel("Speed")
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(2, 1, 2)  # 2 rows, 1 column, second plot
sns.lineplot(x="Block Size", y="Speed", hue="Encryption", data=df[df["Operation"] == "write"], markers=True)
plt.title("Speed vs. Block Size (Write)")
plt.xlabel("Block Size")
plt.ylabel("Speed")
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(2, 1, 1) 
sns.lineplot(x="File Size", y="Speed", hue="Encryption", data=df[df["Operation"] == "read"], markers=True)
plt.title("Speed vs. File Size (Read)")
plt.xlabel("File Size")
plt.ylabel("Speed")
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(2, 1, 1) 
sns.lineplot(x="File Size", y="Speed", hue="Encryption", data=df[df["Operation"] == "write"], markers=True)
plt.title("Speed vs. File Size (Write)")
plt.xlabel("File Size")
plt.ylabel("Speed")
```
